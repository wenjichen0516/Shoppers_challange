```python
def diff_days(s1,s2):
    date_format = "%Y-%m-%d"
    a = datetime.strptime(s1, date_format)
    b = datetime.strptime(s2, date_format)
    return (b-a).days
```

```python
for index,row in df_transaction.iterrows():
    if(row['id'] in df_train_offer['id']):

        if(row['company'] == df_train_offer.loc[row['id'],'company']):
            
            df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company'] += 1
            
            date_diff_days  = diff_days(row['date'],df_train_offer.loc[row['id'],'offerdate'])
            if(date_diff_days < 30):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_30days'] += 1
            elif(30 < date_diff_days < 60):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_60days'] += 1
            elif(60 < date_diff_days < 90):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_90days'] += 1
            elif(90 < date_diff_days < 120):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_120days'] += 1
            elif(120 < date_diff_days < 150):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_150days'] += 1
            elif(150 < date_diff_days < 180):
                df_train_offer.loc[df_train_offer['id'] == row['id'],'have_bought_company_180days'] += 1
```
